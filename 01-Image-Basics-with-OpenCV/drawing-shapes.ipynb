```python
# Drawing
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
blank_img
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(200, 40), pt2=(412,60), color=(0,255,0), thickness=10)

plt.imshow(blank_img)
```

```python
# Circle

cv2.circle(
    img=blank_img,
    center=(200,200),
    radius=50,
    thickness=8,
    color=(0,0,255)
)
plt.imshow(blank_img)
```

```python
cv2.line(
    img=blank_img,
    pt1=(50,50),
    pt2=(400, 510),
    thickness=8,
    color=(0,0,255)
)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_PLAIN
cv2.putText(
    blank_img, 
    text="hello",
    org=(10,500),
    fontFace=font,
    fontScale=4,
    color=(255, 255,255), 
    thickness=3,
    lineType=cv2.LINE_AA
)
plt.imshow(blank_img)
```

```python
blank_img = np.zeros((512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
# Verticies 
verticies = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
pts = verticies.reshape(-1,1,2)
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=4,)


plt.imshow(blank_img)
```
